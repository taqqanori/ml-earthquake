```python
import folium, json

with open('out/tokyo/validations.json') as f:
    validations = json.load(f)

_map = folium.Map(location=[35.680934, 139.767551])

count = 0
for validation in validations['validations']:
    with open('out/tokyo/' + validation['detail']) as f:
        detail = json.load(f)
    color = 'green' if 0 < validation['fact'] and 0.5 <= validation['prediction'] else 'red'
    for eq in detail['earthquakes']:
        count += 1
        folium.CircleMarker(
            location=[eq['latitude'], eq['longitude']],
            radius=0.1,
            color=color,
            fill=True,
            fill_color=color,
            opacity=0.1
        ).add_to(_map)

print(count)

_map
```
